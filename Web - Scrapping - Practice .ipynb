```python
def star_tag_tag(find_all_tag):
    find_star = find_all_tag
    star_tag_count = []
    for i in find_star:
        star_tag_count.append(i.text)
    return star_tag_count
```

```python
def my_func (num):
    strr = num.strip()
    if strr[-1] == 'k':
        return int(float(strr[:-1])*1000)
    return strr
```

```python
my_func(str_tag[0])
```

```python
def my_function_n (name_type, str_tag):
    name_sep = name_type.find_all('a')
    name = name_sep[0].text.strip()
    topics = name_sep[1].text.strip()
    stars = my_func(str_tag)
    url_topic = "https://github.com"+name_sep[1]['href']
    return name, topics, stars, url_topic
```

```python
def url_link (url):
    #get page for scrapping
    response_url = requests.get(url)
    if response_url.status_code == 200:
        #save in new file
        repo_url = response_url.text
        doc = BeautifulSoup(repo_url,'html.parser')
        
```

```python
def direct_url (url):
    response_url = requests.get(url)
    
    if response_url.status_code == 200:
        page_cont = response_url.text
    
    doc = BeautifulSoup(page_cont, 'html.parser')
    return doc
    

def get_repos_info (tag_a, str_tag):
    name_sep = tag_a.find_all('a')
    name = name_sep[0].text.strip()
    topics = name_sep[1].text.strip()
    stars = my_func(str_tag.text.strip())
    url_topic = "https://github.com"+name_sep[1]['href']
    return name, topics, stars, url_topic


def get_basic_tags (doc):
    a_sep = doc.find_all('h3',{'class':'f3 color-fg-muted text-normal lh-condensed'})
    star_tag_all_url = doc.find_all('span', {'class':'Counter js-social-count'})

    my_dit = {
        'name': [],
        'topic': [],
        'stars': [],
        'link': []
    }
    for i in range(len(name_type)):

        info = get_repos_info(a_sep[i], star_tag_all_url[i])
        my_dit['name'].append(info[0])
        my_dit['topic'].append(info[1])
        my_dit['stars'].append(info[2])
        my_dit['link'].append(info[3])
    return pd.DataFrame(my_dit)
```

```python
def topic_name(topic_url):
    response_url = requests.get(topic_url)
    if response_url.status_code == 200:
        page_cont = response_url.text
        soup = BeautifulSoup(page_cont, 'html.parser')
        topics_name_url = soup.find_all('p', {'class': 'f3 lh-condensed mb-0 mt-1 Link--primary'})
        topic_names = []
        for i in topics_name_url:
            topic_names.append(i.text.strip())
        return topic_names


def topic_details(soup):
    topic_details_url= soup.find_all('p',{'class': 'f5 color-fg-muted mb-0 mt-1'})
    topic_details = []
    for y in topic_details_url:
        topic_details.append(y.text.strip())
    return topic_details

    
def topic_url (soup):
    topic_url_url = soup.find_all('a',{'class': 'no-underline flex-1 d-flex flex-column'})
    topic_url = []
    for z in topic_url_url:
        topic_url.append('https://github.com'+z['href'])
    return topic_url 


def dic ():
    github_my_file = {
    'topic_name': topic_name('https://github.com/topics'),
    'topic_details': topic_details(soup),
    'topic_url': topic_url (soup)
    }
    return pd.DataFrame(github_my_file)
```
