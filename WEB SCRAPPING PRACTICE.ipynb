### using Beautiful soup for extracting data from HTML

```python
!pip install beautifulsoup4  --upgrade --quiet
```

```python
from bs4 import BeautifulSoup
```

```python
type(BeautifulSoup)
```

```python
with open ('webpage_url.html', 'r') as f:
    html_source = f.read()
```

```python
html_source[:1000]
```

For checking:

```python
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
type(soup)
```

```python
doc = BeautifulSoup(html_source, 'html.parser')
```

```python
type(doc)
```

```python
doc
```

```python
topic_tags_all = doc.find_all('p', {'class': "f3 lh-condensed mb-0 mt-1 Link--primary"})
```

```python
topic_tags_all
```

```python
len(topic_tags_all)
```

```python
detail_tag_all = doc.find_all('p', {'class': 'f5 color-fg-muted mb-0 mt-1'})
```

```python
detail_tag_all
```

```python
url_tag_all = doc.find_all('a', {'class':"no-underline flex-1 d-flex flex-column"})
```

```python
len(url_tag_all)
```

```python
url_tag_all
```

```python
url_tag_all[0]['href']
```

```python
topic_title = []

for x  in topic_tags_all:
    topic_title.append(x.text)

print(topic_title)
```

```python
url_link = []

for y in url_tag_all:
    url_link.append('https://github.com'+y['href'])
    
print(url_link)
```

```python
tag_details = []

for z in detail_tag_all:
    tag_details.append(z.text.strip())
    
print(tag_details)
```

# Now making Dataframe

```python
data = {
    'Topic_data': topic_title,
    'Topic_url': url_link,
    'Topic_detail': tag_details 
}
github_df = pd.DataFrame(data)
```
